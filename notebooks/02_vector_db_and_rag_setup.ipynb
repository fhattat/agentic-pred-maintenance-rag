## 1.1. RAG Altyapısı: ChromaDB ve Yerel Embedding Kurulumu
Açıklama: Bu bölümde, teknik kılavuzu küçük parçalara (chunks) bölerek yerel bir vektör veritabanına kaydedeceğiz. Bu işlem için tamamen ücretsiz ve yerel çalışan sentence-transformers kütüphanesini kullanacağız.

```python
import os
import chromadb
from chromadb.utils import embedding_functions

# 1. Klasör Kontrolü
if not os.path.exists('../data/db'):
    os.makedirs('../data/db')

# 2. Embedding Fonksiyonu
# Bu kısım yerel işlemcini kullanır, kota harcamaz.
local_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")

# 3. ChromaDB İstemcisi
client = chromadb.PersistentClient(path="../data/db")

# 4. Koleksiyon Oluşturma
collection = client.get_or_create_collection(
    name="technical_manual", 
    embedding_function=local_ef
)

print("✅ Altyapı başarıyla güncellendi ve ChromaDB yüklendi.")
```

```python
# Veritabanında kaç kayıt var?
count = collection.count()
print(f"Veritabanındaki toplam kayıt sayısı: {count}")

# Eğer 0 ise, ekleme adımı başarısız olmuş demektir.
if count > 0:
    print("İlk kayda bir bakalım:")
    print(collection.peek(1))
```

Eğer sayı 0 çıktıysa, önceki hücredeki bölme (split) mantığını daha garanti bir hale getirip tekrar çalıştıralım. Bazen dosya yolları veya karakter kodlamaları sorun çıkarabilir.

```python
import os

# Dosya yolunu tam kontrol edelim
manual_path = '../docs/technical_manual.md'

if os.path.exists(manual_path):
    with open(manual_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Eğer '##' ile bölünmüyorsa, tüm metni tek parça olarak ekleyelim (şimdilik)
    sections = [s.strip() for s in content.split('##') if s.strip()]
    
    if not sections: # Eğer hala boşsa
        sections = [content.strip()]

    documents = sections
    metadatas = [{"source": "manual", "index": i} for i in range(len(sections))]
    ids = [f"id_{i}" for i in range(len(sections))]

    # Koleksiyonu temizleyip tekrar ekleyelim
    # collection.delete(ids=[...]) yerine basitçe ekliyoruz:
    collection.add(
        documents=documents,
        metadatas=metadatas,
        ids=ids
    )
    print(f"✅ Yeniden deneme başarılı! {len(documents)} parça eklendi.")
else:
    print(f"❌ HATA: {manual_path} dosyası bulunamadı! Lütfen yolu kontrol et.")
```

```python
import os

# Dosya yolunu kontrol et
file_path = '../docs/technical_manual.md'

if os.path.exists(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        full_text = f.read()
    
    # 1. Strateji: '##' ile bölmeyi dene
    chunks = [c.strip() for c in full_text.split('##') if len(c.strip()) > 10]
    
    # 2. Strateji: Eğer '##' yoksa paragraflara (çift satır başı) göre böl
    if len(chunks) <= 1:
        chunks = [p.strip() for p in full_text.split('\n\n') if len(p.strip()) > 10]
    
    # 3. Strateji: Hiçbiri olmazsa tek parça al
    if not chunks:
        chunks = [full_text.strip()]

    # Veritabanına ekle
    doc_ids = [f"id_{i}" for i in range(len(chunks))]
    metadatas = [{"source": "manual", "part": i} for i in range(len(chunks))]
    
    collection.add(
        documents=chunks,
        ids=doc_ids,
        metadatas=metadatas
    )
    print(f"✅ Başarılı! {len(chunks)} parça veritabanına eklendi.")
    print(f"Güncel toplam kayıt sayısı: {collection.count()}")
else:
    print("❌ HATA: '../docs/technical_manual.md' dosyası bulunamadı!")
```

##  Vektör Arama Testi (Semantic Search)

```python
# Mühendis sorusu
test_query = "What happens if vibration exceeds 1.3 and temperature is high?"

# Arama
results = collection.query(
    query_texts=[test_query],
    n_results=2
)

print("--- RAG ARAMA SONUÇLARI ---")
if results['documents'][0]:
    for i, doc in enumerate(results['documents'][0]):
        print(f"\n[Eşleşen Bilgi {i+1}]:\n{doc}")
        print("-" * 20)
else:
    print("Arama sonucu boş döndü. Lütfen 1. adımı tekrar kontrol et.")
```

**Mühendislik Notu**

Şu an yaptığımız şey, bir kütüphanecinin rafları düzenlemesi gibidir.

Embedding: Kitapların konularını (anlamlarını) anlamak.

Vector DB (Chroma): Kitapları konularına göre raflara dizmek.

Query: İlgili konudaki en yakın kitabı saniyeler içinde raftan çekmek.

Bu kodları çalıştırdığında metinleri görebiliyor musun? Eğer görebiliyorsan, projenin "Retrieval" (Bilgi Getirme) motoru mükemmel çalışıyor demektir.

# Entegre RAG Fonksiyonu ve Prompt Tasarımı

Şimdi, veritabanından gelen teknik metinleri ve makine verilerini birleştirip "nihai raporu" oluşturacak fonksiyonu yazacağız. Bu fonksiyon, AI Engineering'deki "Prompt Injection" ve "Context Management" süreçlerini temsil eder.

## Akıllı RAG Ajanı: Veri ve Rehberi Birleştirme

```python
def get_diagnosis_report(sensor_data, model):
    """
    Bu fonksiyon hem veritabanından bilgi çeker hem de LLM ile analiz yapar (Not: Bu fonksiyon Notebook 03'te Multi-Agent sistemiyle değiştirilmiştir).
    """
    # 1. Veritabanında Arama Yap (Retrieve)
    # Sorguyu sensör verilerindeki kritik değerlere göre şekillendiriyoruz
    query = f"Vibration: {sensor_data['vibration']}, Temp: {sensor_data['temperature']}, Current: {sensor_data['current']}"
    results = collection.query(query_texts=[query], n_results=2)
    
    retrieved_docs = "\n".join(results['documents'][0])
    
    # 2. Prompt Tasarımı (Context Injection)
    final_prompt = f"""
    Sen uzman bir Endüstriyel Bakım Mühendisisin. Aşağıdaki verileri ve teknik kılavuz parçalarını kullanarak bir teşhis raporu hazırla.

    [GÜNCEL MAKİNE VERİLERİ]:
    {sensor_data}

    [TEKNİK KILAVUZDAN İLGİLİ BÖLÜMLER]:
    {retrieved_docs}

    GÖREV:
    1. Makinenin şu anki durumunu (Normal/Uyarı/Kritik) belirle.
    2. Kılavuzdaki eşik değerlerine göre hangi parametrelerin sorunlu olduğunu açıkla.
    3. Eğer bir arıza riski varsa, kılavuzdaki müdahale adımlarını öner.

    Lütfen profesyonel ve kısa bir rapor sun.
    """
    
    # 3. Üretim (Generate) - Kota dostu hata yönetimiyle
    try:
        # Sorgu öncesi kısa bir bekleme (Kota koruması)
        import time
        time.sleep(1) 
        
        response = model.generate_content(final_prompt)
        return response.text
    except Exception as e:
        return f"Şu an rapor üretilemiyor (Kota/Bağlantı Sorunu): {str(e)}"

print("✅ Entegre RAG fonksiyonu hazır.")
```
